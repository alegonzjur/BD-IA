2. Para los números del -3 al 3, calcula la función seno.

```python
x = np.arange(-3,3)
y = np.sin(x)


fig, ax = plt.subplots(figsize = (10,6))
ax.plot(x,y)
ax.set(xlabel='Valores X', ylabel='Sen (x)', title='Seno de valores');
```

3. Modificar el ejercicio anterior para que muestre simultaneamente la funcion seno y coseno.

```python
z = np.cos(x)
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(x,y)
ax.plot(x,z)
ax.set(xlabel='Valores X', ylabel='Valores Y', title='Valores');
```
